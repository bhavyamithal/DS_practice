# LR, RR & Lasso Regression

```python
# House pricing dataset
# from sklearn.datasets import load_boston
from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
housing = fetch_california_housing()
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
dataset.head()
```

```python
dataset['Price'] = housing.target
# dataset = dataset.drop(['target'], axis=1)
dataset.head()
```

        Dividing the dataset into dependent and independent features

```python
x = dataset.iloc[:, :-1] # independent variables
y = dataset.iloc[:, -1] # dependent variables
x.head()
```

## Linear Regression


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
mse = cross_val_score(lin_reg, x_train, y_train, scoring = 'neg_mean_squared_error', cv = 5)
mean_mse = np.mean(mse)
mean_mse
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-2, 1, 5, 10, 20, 30, 35, 100]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring = 'neg_mean_squared_error', cv = 5)
ridge_regressor.fit(x_train, y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4,1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 100]}
lasso_regressor = GridSearchCV(lasso, parameters, scoring = 'neg_mean_squared_error', cv = 5)
lasso_regressor.fit(x_train, y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred = lin_reg.predict(x_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred, y_test)
print(r2_score1)

y_pred = ridge_regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred, y_test)
print(r2_score1)

y_pred = lasso_regressor.predict(x_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred, y_test)
print(r2_score1)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
# Independent Features
x = pd.DataFrame(df['data'], columns=df['feature_names'])
x
```

```python
# Dependent Feature
y = pd.DataFrame(df['target'], columns=['Target'])
y
```

                        Check if y is imbalanced or not

```python
y['Target'].value_counts()

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, 
                                                    test_size=.2, 
                                                    random_state=42)
parameters = [{'C': [1, 5, 10]}, {'max_iter': [100, 150]}]
log_reg = LogisticRegression(C=100, max_iter=100)
model = GridSearchCV(log_reg, param_grid=parameters, cv=5, scoring='f1')
model.fit(x_train, y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
y_pred = model.predict(x_test)
y_pred
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
