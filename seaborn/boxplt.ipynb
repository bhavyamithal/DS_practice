```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = sns.load_dataset('iris')
df
```

```python
sns.boxplot(x='species', y='sepal_length', data=df, showmeans=True, 
            meanprops={'marker':'o', 'markerfacecolor':'white', 'markeredgecolor':'black'}, 
            palette='Blues')
plt.show()
```
