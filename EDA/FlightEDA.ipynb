# Flight Price Prediction(EDA + Feature Engineering)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('FlightData/Data_Train.xlsx')
df_train.head()
```

```python
df_test = pd.read_excel('FlightData/Test_set.xlsx')
df_test.head()
```

```python
final_df = pd.concat([df_train, df_test], axis=0)
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head()
```

```python
# final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x: x.split('/')[0])
# final_df['Month'] = final_df['Date_of_Journey'].apply(lambda x: x.split('/')[1])
# final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x: x.split('/')[2])
# final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
# final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute'] = final_df['Arrival_Time'].str.split(':').str[1]
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df['Dept_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_Minute'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_Hour'] = final_df['Dept_Hour'].astype(int)
final_df['Dept_Minute'] = final_df['Dept_Minute'].astype(int)

final_df.info()
```

```python
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, 
                                                       '2 stops': 2, '3 stops': 3, 
                                                       '4 stops': 4, 'nan': 1})
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.Additional_Info.value_counts()
```

```python
final_df.head()
```

```python
# final_df[final_df['Duration_Hour']=='5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_Hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)*60
final_df['Duration_Minute'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0].replace(np.nan, 0)
```

```python
final_df['Duration_Minute'] = final_df['Duration_Minute'].astype(int)
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)
```

```python
final_df['Duration'] = final_df['Duration_Hour'] + final_df['Duration_Minute']
```

```python
final_df.drop(['Duration_Hour', 'Duration_Minute'], axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.Airline.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
final_df['Airline'] = le.fit_transform(final_df['Airline'])
final_df['Source'] = le.fit_transform(final_df['Source'])
final_df['Destination'] = le.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])

final_df.head()
```
